```python
dataset = pd.read_excel('Lab Session2 Data.xlsx')
dataset['Tokens']=dataset['Report'].apply(nl.word_tokenize)
dataset
```

```python
#A1-A2
from nltk.corpus.reader.tagged import word_tokenize
from nltk.corpus import stopwords
token = []
for dataset in dataset['Report']:
  token.append(word_tokenize(dataset))
print(token)
import itertools
flat = list(itertools.chain(*token))
flat = [name.lower() for name in flat]
pop_tok = [*set(flat)]
print(pop_tok)
print(len(pop_tok))
```

```python
#A3-A4
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
filtered_sentence = []
for w in pop_tok:
    if w not in stop_words:
        filtered_sentence.append(w)
print(filtered_sentence)
print(len(filtered_sentence))
```
